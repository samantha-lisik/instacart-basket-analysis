8. Missing values in the df_ords dataframe

```python
df_ords.isnull().sum()
```

Findings:
Only one column has missing values: days_since_last_order → 206,209 missing values
All other columns (order_id, user_id, eval_set, order_number, orders_day_of_week, order_hour_of_day) have 0 missing values.

Explanation:
'days_since_last_order' measures the number of days since a user’s previous order.

For a user’s first order, there is no previous order, so this value is naturally missing (NaN).

This is expected behaviour and not a data quality issue.
