```python
# Import libraries
import pandas as pd
import os
```

3. Merge dataframes

```python
# Merge df_ords_prior with df_ords
df_merged = df_ords.merge(df_ords_prior, on='order_id', indicator=True)
```

```python
df_merged
```

4. Create a new dataframe 'df_merged_large' that contains the combined df_ords and df_ords_prior dataframes and uses the “order_id” column as its key

```python
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)
```

```python
df_merged_large.head()
```

```python
# Check for full match
df_merged_large['_merge'].value_counts()
```
