1. Import libraries and orders_products_combined dataframe

```python
# Import libraries
import pandas as pd
import os
```

```python
# Define path:
path = r'/Users/samlisik/Documents/Instacart Basket Analysis'
```

```python
# Import the orders_products_combined dataframe
df_ords_prods_combined = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

```python
# Check the shape of the imported dataframe
df_ords_prods_combined.shape
```

2. Combine the orders_products_combined and the products_clean dataframes

```python
# Import cleaned products dataset
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_clean.csv'))
```

```python
# Merge orders_products_combined with products_clean on 'product_id'
df_ords_prods_merged = pd.merge(df_ords_prods_combined, df_prods, on='product_id', how='left')
```

```python
# Check shape
df_ords_prods_merged.shape
```

```python
df_ords_prods_merged.head()
```

```python
# Confirm the results using the merge flag
df_ords_prods_merged['_merge'].value_counts()
```

3. Export the newly created dataframe

```python
# Export the merged dataframe as a Pickle file (due to size)
df_ords_prods_merged.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge.pkl'))
```
