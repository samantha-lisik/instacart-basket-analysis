4. Line charts

```python
# Create a list holding True/Flase values to the test np.random.rant() <= 0.7
np.random.seed(4)
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
dev
```

```python
np.random.rand(10)
```

```python
# Store 70% of the sample in the dataframe big
big = ords_prods_merge[dev]
```

```python
# Store 30% of the sample in the dataframe small
small = ords_prods_merge[~dev]
```

```python
# Check if the subsetting procedure was successful
len(ords_prods_merge)
```

```python
len(big) + len(small)
```

```python
# Reduce samples to only those columns necessary for the chart
df_2 = small[['orders_day_of_week','prices']]
```

```python
# Create chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Export line chart
line.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'line_prices_dow.png'))
```
