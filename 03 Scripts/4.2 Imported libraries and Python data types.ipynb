2. Working with Python data types

```python
# Method 1: Direct addition of integers
result1 = 50 + 50
print(result1)
```

```python
# Method 2: Combination of addition and subtraction
result2 = 150 - 25 - 25
print(result2)
```

```python
# Method 3: Using variables and a mix of operations
x = 200
y = 75
z = 25
result3 = x - y - z
print(result3)
```

```python
# Define two floating-point variables
a = 25.5
b = 5.1

# Divide them
result = a / b

# Print the result
print(result)
```

```python
# Define each character as a string
s = 's'
n = 'n'
a = 'a'
k = 'k'
e = 'e'

# Concatenate them into one word
word = s + n + a + k + e

# Print the result
print(word)
```

```python
# First sentence
the = "The "
snake = "snake "
hides = "hides "
under = "under "
rocks = "rocks."

sentence1 = the + snake + hides + under + rocks

# Second sentence
a = "A "
green = "green "
snake2 = "snake "
slithers = "slithers "
away = "away."

sentence2 = a + green + snake2 + slithers + away

# Print results
print(sentence1)
print(sentence2)
```
